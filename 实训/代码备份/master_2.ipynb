```python
import pandas as pd
import jieba
import numpy as np
import os
import synonyms
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
data = pd.read_csv('D:\\learn\\小学期-3\\eg\\data.csv', encoding = 'utf-8')
data.head()
```

```python
content = np.array(data['content'])
content
```

```python
#结巴分词
List_word = []
for s in content:
    str = ' '.join(jieba.cut(s))
    List_word.append( str )
List_word
```

```python
#去除停用词
with open('D:\\learn\\小学期-3\\eg\\stop_words_zh.txt', 'r', encoding='utf-8-sig') as fs:
    stop_word = fs.read()

stop_words = stop_word.split('\n')

List_w = []
for s in List_word:
    mid_s = s.split(' ')
    mid_w = ''
    for word in mid_s:
        if not(word in stop_words) and len(word) > 1:
            mid_w = mid_w + word +' '
    mid_w = mid_w[0:-1]
    List_w.append(mid_w)
List_w
```

```python
corpos = pd.DataFrame({'content':List_w, 'style':data['style']})
corpos.head()
```

```python
#将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频
countVectorizer = CountVectorizer(min_df = 0, token_pattern=u'(?u)\\b[^\\d\\W]\\w+\\b')#增min_df=0参数，保留最小长度为0的分词
textVector = countVectorizer.fit_transform(corpos['content'])
df_vector = pd.DataFrame(textVector.toarray(), columns = countVectorizer.get_feature_names())

#计算每个词语的tf-idf权值
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(textVector)
df_tfidf = pd.DataFrame(tfidf.toarray())

#tf-idf值越大，关键字的权重越大，对所有的行进行排序，取前5个关键字
sort = np.argsort(tfidf.toarray(), axis = 1)[:, -5:]
names = countVectorizer.get_feature_names()
keyWords = pd.Index(names)[sort].values
keyWords
```

```python
#所有文档的前五个关键词
df_keys = pd.DataFrame({'content':data['content'], 
                             'style':data['style'],
                             'key1':keyWords[:,0],
                             'key2':keyWords[:,1],
                             'key3':keyWords[:,2],
                             'key4':keyWords[:,3],
                             'key5':keyWords[:,4],
                             })
df_keys.head()
```

```python
'''生成3000字草稿'''
style = 1 #风格
keyWords = ['金融', '科技', '融资', '创新'] #关键词

df_source = df_keys
df = df_source[df_source['style'] == 1]
df.describe()
```

```python
a = np.array(df)
```

```python
fre = []
for w in keyWords:
    t = []
    for i in a:
        similiar = 0
        for j in range(2, 7):
            similiar = similiar + synonyms.compare(w, i[j])
        t.append(similiar)
    fre.append(t)
fre   
```

```python
a[0][0]
```

```python
article = ''
while len(article) < 3000:
    for i in range(4):
        sub = fre[i].index(max(fre[i]))
        article = article + a[sub][0]
        fre[i][sub] = 0
with open('D:\\learn\\小学期-3\\eg\\article.txt', 'w', encoding = 'utf-8-sig') as w_f:
    w_f.write(article)       
```
